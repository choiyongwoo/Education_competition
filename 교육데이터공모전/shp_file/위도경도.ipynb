```python
# 위도 경도(튜플형식)
def addr_to_lat_lon(addr):
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query={address}'.format(address=addr)
    headers = {"Authorization": "KakaoAK " + api_key}
    try:
        result = json.loads(str(requests.get(url, headers=headers).text))
        match_first = result['documents'][0]['address']
        return float(match_first['x']), float(match_first['y'])
    except:
        return None
```

```python
def tuple_to_text(df, column_name):
    try:
        # 튜플 형태의 좌표를 텍스트 형태로 변환하는 함수
        def tuple_to_str(coordinates):
            try:
                return str(coordinates)
            except (ValueError, TypeError):
                return None

        # 데이터프레임의 해당 열을 순회하면서 텍스트 형태로 변환하여 새로운 열 생성
        df[column_name + '_text'] = df[column_name].apply(tuple_to_str)
        return df
    except KeyError:
        print("해당 열이 존재하지 않습니다.")
        return None
```

```python
# (126.503945566825, 37.7633975743133)

# text = '(126.503945566825, 37.7633975743133)'
def split_lat(lat_lon):
    try:
        lat = float( lat_lon.strip('()').split(',')[0])
        return lat
    except KeyError:
        return None
# split_lat(text)
```

```python
def split_lon(lat_lon):
    try:
        lon = float( lat_lon.strip('()').split(',')[1])
        return lon
    except KeyError:
        return None
```
