### 후진소거법

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols

# 설명 변수 리스트
variables = ['총학생수','대중교통/도보(분)','학생1인당교지면적','학원수','인구수대비학생수']

# 반응 변수
y = scaled_df['학교수']
selected_variables = variables.copy()  # 초기에는 모든 변수가 선택된 상태
sl_remove = 0.1

# 각 스텝별로 선택된 변수들, 수정된 결정계수, 스텝을 저장할 리스트
sv_per_step = []
adjusted_r_squared = []
steps = []
step = 0

while len(selected_variables) > 0:
    # 현재 선택된 변수들로 OLS 모델을 적합
    X = sm.add_constant(scaled_df[selected_variables])
    model = sm.OLS(y, X).fit()
    p_vals = model.pvalues[1:]  # 절편항의 p-value는 제외

    max_pval = p_vals.max()  # 최대 p-value
    if max_pval >= sl_remove:  # 최대 p-value값이 기준값보다 크거나 같으면 제외
        remove_variable = p_vals.idxmax()
        selected_variables.remove(remove_variable)

        step += 1
        steps.append(step)
        adj_r_squared = sm.OLS(y, sm.add_constant(scaled_df[selected_variables])).fit().rsquared_adj
        adjusted_r_squared.append(adj_r_squared)
        sv_per_step.append(selected_variables.copy())
    else:
        break

# 그래프 시각화
fig = plt.figure(figsize=(10, 10))
fig.set_facecolor('white')

font_size = 15
plt.xticks(steps, [f'step {s}\n' + '\n'.join(sv_per_step[i]) for i, s in enumerate(steps)], fontsize=12)
plt.plot(steps, adjusted_r_squared, marker='o')

plt.ylabel('Adjusted R Squared', fontsize=font_size)
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

import numpy as np
%matplotlib inline

kmenas= KMeans(n_clusters=4, init='k-means++', max_iter=300, random_state=42)
kmenas.fit(cluster_df[['총학생수','대중교통/도보(분)','학생1인당교지면적']])

cluster_df['cluster']=kmenas.labels_

pca = PCA(n_components=2)
pca_transformed = pca.fit_transform(cluster_df[['총학생수','대중교통/도보(분)','학생1인당교지면적']])

cluster_df['pca_x'] = pca_transformed[:,0]
cluster_df['pca_y'] = pca_transformed[:,1]

# cluster 값이 0, 1, 2 인 경우마다 별도의 Index로 추출
marker0_ind = cluster_df[cluster_df['cluster']==0].index
marker1_ind = cluster_df[cluster_df['cluster']==1].index
marker2_ind = cluster_df[cluster_df['cluster']==2].index
marker3_ind = cluster_df[cluster_df['cluster']==3].index


plt.subplot(1,2,1)
plt.title('target', fontsize=20)

plot_df1 = cluster_df[cluster_df['target']==0]
plot_df2 = cluster_df[cluster_df['target']==1]

plt.scatter(plot_df1['pca_x'], plot_df1['pca_y'], c="silver", s=60)
plt.scatter(plot_df2['pca_x'], plot_df2['pca_y'], c="tomato", s=60)

plt.subplot(1,2,2)
# cluster값 0, 1, 2에 해당하는 Index로 각 cluster 레벨의 pca_x, pca_y 값 추출. o, s, ^ 로 marker 표시
plt.scatter(x=cluster_df.loc[marker0_ind,'pca_x'], y=cluster_df.loc[marker0_ind,'pca_y'], marker='o') 
plt.scatter(x=cluster_df.loc[marker1_ind,'pca_x'], y=cluster_df.loc[marker1_ind,'pca_y'], marker='s')
plt.scatter(x=cluster_df.loc[marker2_ind,'pca_x'], y=cluster_df.loc[marker2_ind,'pca_y'], marker='^')
plt.scatter(x=cluster_df.loc[marker3_ind,'pca_x'], y=cluster_df.loc[marker3_ind,'pca_y'])


plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('3 Clusters Visualization by 2 PCA Components')
plt.show()
```

```python
import numpy as np
from sklearn.cluster import MeanShift, estimate_bandwidth
import matplotlib.pyplot as plt

# 임의의 데이터 생성

X=cluster_df[['총학생수','대중교통/도보(분)','학생1인당교지면적']]

pca = PCA(n_components=2)
pca_transformed = pca.fit_transform(cluster_df[['총학생수','대중교통/도보(분)','학생1인당교지면적']])

cluster_df['pca_x'] = pca_transformed[:,0]
cluster_df['pca_y'] = pca_transformed[:,1]

# 대역폭 추정
bandwidth = estimate_bandwidth(X, quantile=0.2, n_samples=None)

# Mean Shift 모델 생성 및 학습
ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
ms.fit(X)

# 클러스터 중심 및 레이블 얻기
cluster_centers = ms.cluster_centers_
labels = ms.labels_

# 고유한 레이블 개수 찾기 (클러스터의 수)
n_clusters_ = len(np.unique(labels))
cluster_df['cluster']= labels

# 결과 출력
print("클러스터의 수: {}".format(n_clusters_))

# 클러스터링 결과 시각화
plt.scatter(cluster_df['pca_x'], cluster_df['pca_y'], c=labels, marker='o')
# plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', s=100, alpha=0.5)
plt.title('Mean Shift Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
